```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
df['TotalPremium'].hist()
df['Gender'].value_counts().plot(kind='bar')
```

```python
df[['TotalPremium', 'TotalClaims']].corr()
sns.scatterplot(x='TotalPremium', y='TotalClaims', hue='PostalCode', data=df)
```

```python
sns.boxplot(x=df['TotalPremium'])
```

```python
# Compare premium and claims across provinces with a bar or heatmap
plot = sns.barplot(x='Province', y='TotalPremium', data=df)


plot.set_xticklabels(plot.get_xticklabels(), rotation=45)



```

```python
plot = sns.violinplot(x='PostalCode', y='TotalPremium', data=df)

plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
```
